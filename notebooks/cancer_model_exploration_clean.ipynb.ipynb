### Model Building and Evaluation Using Logistic Regression


```python

# DEFINE FEATURES (X) AND TARGET (y)
X = df[
    [
        'Age',
        'Gender',
        'BMI',
        'Smoking',
        'GeneticRisk',
        'PhysicalActivity',
        'AlcoholIntake',
        'CancerHistory']]

y = df['Diagnosis'] 

##  FEATURE SCALING
scaler = StandardScaler()             
X_scaled = scaler.fit_transform(X)   

## TRAIN-TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


## TRAIN LOGISTIC REGRESSION MODEL
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

## PREDICTION ON TEST DATA
y_pred = model.predict(X_test)

## EVALUATION
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

### MODEL AND GUI

```python

## IMPORT LIBRARIES
import pandas as pd
import tkinter as tk
from tkinter import messagebox

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, f1_score

## LOADING DATASET
df = pd.read_csv("C:/Users/Samrath Narula/Downloads/The_Cancer_data_1500_V2.csv")  

## Remove rows with missing values
df.dropna(inplace=True)

## DEFINING FEATURES (X) AND TARGET (y)
X = df[
    [
        'Age',
        'Gender',
        'BMI',
        'Smoking',
        'GeneticRisk',
        'PhysicalActivity',
        'AlcoholIntake',
        'CancerHistory']]

y = df['Diagnosis'] 

##  FEATURE SCALING
scaler = StandardScaler()             
X_scaled = scaler.fit_transform(X)   

##  TRAIN-TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

## TRAIN LOGISTIC REGRESSION MODEL
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

## PREDICTION ON TEST DATA
y_pred = model.predict(X_test)

## EVALUATION
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
print("Accuracy of the model :", accuracy_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

### GUI ########
def predict_from_gui():                    ## Take user input from GUI fields
    try:
        
        user_input = pd.DataFrame([{
            'Age': int(age.get()),
            'Gender': int(gender.get()),
            'BMI': float(bmi.get()),
            'Smoking': int(smoking.get()),
            'GeneticRisk': int(genetic.get()),
            'PhysicalActivity': float(activity.get()),
            'AlcoholIntake': float(alcohol.get()),
            'CancerHistory': int(history.get()) }])

                                                     ## Scale user input using SAME scaler
        user_scaled = scaler.transform(user_input)

                                                    ## Predict using trained model
        prediction = model.predict(user_scaled)
        probability = model.predict_proba(user_scaled)[0][1] * 100

                                ## Show result in message box
        if prediction[0] == 1:
            result = f"⚠ CANCER DETECTED\nRisk: {round(probability)}%"

        else:
            result = f"⚠ CANCER DETECTED\nRisk: {round(probability)}%"


        messagebox.showinfo("Prediction Result", result)

    except:
                                    ## For Error handling 
        messagebox.showerror("Error", "Please enter valid numeric values or numbers")  


## CREATE GUI Interface

root = tk.Tk()
root.title("Cancer Risk Prediction System")
root.geometry("400x820")

BG_COLOR = "brown"
LABEL_COLOR = "red"
ENTRY_BG = "Red"
ENTRY_FG = "black"
FONT_LABEL = ("Arial", 11)
FONT_ENTRY = ("Arial", 11)

root.configure(bg=BG_COLOR)

## GUI INPUT Boxes

tk.Label(root, text="Age").pack()
age = tk.Entry(root); age.pack()

tk.Label(root, text="Gender (0=Female, 1=Male)").pack()
gender = tk.Entry(root); gender.pack()

tk.Label(root, text="BMI").pack()
bmi = tk.Entry(root); bmi.pack()

tk.Label(root, text="Smoking (0=No, 1=Yes)").pack()
smoking = tk.Entry(root); smoking.pack()

tk.Label(root, text="Genetic Risk (0=Low, 1=Medium, 2=High)").pack()
genetic = tk.Entry(root); genetic.pack()

tk.Label(root, text="Physical Activity (hrs/week)").pack()
activity = tk.Entry(root); activity.pack()

tk.Label(root, text="Alcohol Intake").pack()
alcohol = tk.Entry(root); alcohol.pack()

tk.Label(root, text="Family Cancer History (0=No, 1=Yes)").pack()
history = tk.Entry(root); history.pack()

## PREDICT BUTTON

tk.Button(
    root,
    text="Predict Cancer Risk",
    command=predict_from_gui).pack(pady=20)


root.mainloop()
```
